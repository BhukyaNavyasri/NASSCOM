```python
# 🔹 1. DEFINING A BASIC FUNCTION

# A function is a block of code that performs a task and can be reused.

def greet():
    print("Hello! Welcome to Python Functions.")

# Calling the function
greet()
```

```python
def greet_user(name):
    print(f"Hello, {name}!")

greet_user("navya sri")
greet_user("ishu")
```

```python
# 🔹 3. FUNCTION WITH DEFAULT PARAMETERS

def greet_user(name="Guest"):
    print(f"Welcome, {name}!")

greet_user()
greet_user("navya sri")
```

```python
def print_names(*names):
    print("Names passed:", names)
    for name in names:
        print("Hello", name)

print_names("navya sri", "ishu", "kavya")
```

```python
def print_info(**info):
    print("User Info:")
    for key, value in info.items():
        print(f"{key} = {value}")

print_info(name="navya sri", age=18, city="hyderabad")
```

```python
def show_details(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show_details(1, 2, 3, name="Ravi", city="Mumbai")
```

```python
def multiply(a, b):
    return a * b

def square(n):
    return multiply(n, n)

print("Square of 4:", square(4))
```
