```python
# Integer
a = 10
print(type(a))

# Float
b = 3.14
print(type(b))

# String
c = "Python"
print(type(c))

# Boolean
d = True
print(type(d))

# List
e = [1, 2, 3, "four"]
print(type(e), e)

# Dictionary
f = {"name": "Alice", "age": 25}
print(type(f), f)
```
